```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
test = pd.read_excel("D:/Inceptez Technologies/Machine Hackathon/Participants_Data_Final/Data_Test.xlsx")
train = pd.read_excel("D:/Inceptez Technologies/Machine Hackathon/Participants_Data_Final/Data_Train.xlsx")
submission = pd.read_excel("D:/Inceptez Technologies/Machine Hackathon/Participants_Data_Final/Sample_submission.xlsx")
```

```python
train["source"] = "train"
test["source"] = "test"
df = pd.concat([train,test])
df.info()
```

```python
df = df.drop(["CUISINES",'CITY', 'LOCALITY', 'TIME', 'TITLE'],axis=1)
```

```python
df.head()
```

```python
df['VOTES'] = df['VOTES'].astype(str)
df['VOTES'] = df.VOTES.apply(lambda x : x[0:-6])
```

```python
df["VOTES"] = pd.to_numeric(df["VOTES"], errors='coerce')
```

```python
df['VOTES'] = df['VOTES'].replace(np.nan, 0)
```

```python
df['VOTES'] = df['VOTES'].astype('Int64')
```

```python
df.info()
```

```python
df["RATING"] = pd.to_numeric(df["RATING"], errors='coerce')
df["RATING"]=df.RATING.round(0)
```

```python
df["RATING"]=df.RATING.fillna(0)
```

```python
df["RATING"]= df['RATING'].astype('Int64')
```

```python
df["VOTES"]=df.VOTES.transform(lambda x : x.fillna(x.median()))
```

```python
df = df.drop(['RESTAURANT_ID'],axis=1)
df
```

```python
train_final = df[df.source=="train"]
test_final = df[df.source=="test"]
```

```python
train_final = train_final.drop(["source"],axis=1)
train_final.shape
```

```python
test_final = test_final.drop(["source"],axis=1)
test_final.shape
```

```python
test_final = test_final.drop(columns=["COST"])
test_final.info()
```

```python
train_X = train_final.drop(columns=["COST"])
```

```python
train_Y = train_final["COST"]
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
train_X.info()
```

```python
model.fit(train_X, train_Y)
```

```python
dtrain_predictions = model.predict(train_X)
```

```python
from sklearn import metrics
#Print model report:
print("\nModel Report")
print("RMSE : %.4g" % np.sqrt(metrics.mean_squared_error(train_Y.values, dtrain_predictions)))
    
#Predict on testing data:
test_final["COST"] =  model.predict(test_final)
```

```python
Linear_submission = test_final[["COST"]]
Linear_submission.to_excel("D:\Inceptez Technologies\Machine Hackathon\Participants_Data_Final\LinearRegression1.xlsx", encoding='utf8',index=False)
```
